**FILTROS BASICOS PARA VARIABLES CUANTITATIVAS**

```python
#Filtro por comparacion
filtro6=df[df["independence_year"] > 1900]
filtro6
```

**FILTROS BASICOS CON CONETORES LOGICOS**

```python
#Filtro Y
filtro7=df[(df["independence_year"] > 1900) & (df["region"] == "Central Africa")]
filtro7
```

```python
#Filtro o
filtro8=df[(df["independence_year"] > 1900) | (df["region"] == "Central Africa")]
filtro8
```

```python
#Filtro not
filtro9=df[~(df["independence_year"] >= 1900) & ~(df["region"] == "Central Africa")]
filtro9
```

```python
#Convertir archivo filtrado a CSV
filtro9.to_csv("Entregable1.csv")
```
